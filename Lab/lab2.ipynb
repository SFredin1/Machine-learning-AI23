```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Ladda in dataset
cardio = pd.read_csv("../Data/cardio_train.csv", sep=";")

# Dela upp datan i tränings- och testdata
X = cardio.drop(columns=["cardio"], axis=1)
y = cardio["cardio"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definiera pipeline för SVM-modellen
pipe_svm = Pipeline([('scaler', StandardScaler()), ('SVM', SVC())])

# Definiera parametrar som ska optimeras
param_grid_svm = {
    'SVM__C': [0.1, 1, 10, 100],
    'SVM__gamma': [1, 0.1, 0.01, 0.001],
    'SVM__kernel': ['rbf', 'linear']
}

# Använd GridSearchCV för att optimera hyperparametrar för SVM
grid_svm = GridSearchCV(pipe_svm, param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train, y_train)

# Skapa en tränad modell på hela datasetet med de bästa parametrarna för SVM
final_model_svm = grid_svm.best_estimator_
final_model_svm.fit(X, y)

# Spara den tränade SVM-modellen i en .pkl-fil med komprimering
joblib.dump(final_model_svm, 'final_model_svm.pkl', compress=True)

# Definiera pipeline för Random Forest-modellen
pipe_rf = Pipeline([('scaler', StandardScaler()), ('RF', RandomForestClassifier())])

# Definiera parametrar som ska optimeras
param_grid_rf = {
    'RF__n_estimators': [100, 200, 300],
    'RF__max_depth': [None, 10, 20],
}

# Använd GridSearchCV för att optimera hyperparametrar för Random Forest
grid_rf = GridSearchCV(pipe_rf, param_grid_rf, cv=5, scoring='accuracy')
grid_rf.fit(X_train, y_train)

# Skapa en tränad modell på hela datasetet med de bästa parametrarna för Random Forest
final_model_rf = grid_rf.best_estimator_
final_model_rf.fit(X, y)

# Spara den tränade Random Forest-modellen i en .pkl-fil med komprimering
joblib.dump(final_model_rf, 'final_model_rf.pkl', compress=True)

# Utvärdera den slutgiltiga SVM-modellen med valideringsdata
y_pred_svm = final_model_svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
class_rep_svm = classification_report(y_test, y_pred_svm)
print("Accuracy for SVM:", accuracy_svm)
print("Confusion matrix for SVM:\n", conf_matrix_svm)
print("Classification report for SVM:\n", class_rep_svm)

# Utvärdera den slutgiltiga Random Forest-modellen med valideringsdata
y_pred_rf = final_model_rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
class_rep_rf = classification_report(y_test, y_pred_rf)
print("Accuracy for Random Forest:", accuracy_rf)
print("Confusion matrix for Random Forest:\n", conf_matrix_rf)
print("Classification report for Random Forest:\n", class_rep_rf)
```
