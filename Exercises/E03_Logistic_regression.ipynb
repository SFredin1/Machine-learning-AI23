```python
# f) .2

import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target variable to the DataFrame
iris_df['species_name'] = iris.target_names[iris.target]

# Function to remove outliers using Tukey's rule and print information
def remove_outliers(data, columns):
    outlier_rows = []
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        outlier_rows.extend(outliers.index.tolist())
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data, outlier_rows

# Specify columns for outlier removal (excluding the 'species_name' column)
columns_to_remove_outliers = iris.feature_names

# Remove outliers and get outlier rows
iris_df_no_outliers, outlier_rows = remove_outliers(iris_df, columns_to_remove_outliers)

# Display the outlier rows
print("Outlier rows are:", outlier_rows)

# Display the value counts of the 'species_name' column
print("\nValue counts:\n")
print(iris_df_no_outliers['species_name'].value_counts())
```

```python
# 1. Split & scale data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Splitting the data into train and test sets 
X_train, X_test, y_train, y_test = train_test_split(iris_df_no_outliers.drop('species_name', axis=1), iris_df_no_outliers['species_name'], test_size=0.33, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Calculate mean & standard deviation on training data
X_train_mean = X_train_scaled.mean(axis=0)
X_train_std = X_train_scaled.std(axis=0)

# Calculate mean & standard deviation on testing data
X_test_mean = X_test_scaled.mean(axis=0)
X_test_std = X_test_scaled.std(axis=0)

print("Mean & standard deviation on training data: ")
for i, feature in enumerate(iris.feature_names):
    print(f"{feature}: Mean={X_train_mean[i]:.3f}, Std={X_train_std[i]:.3f}")

print("Mean & standard deviation on testing data: ")
for i, feature in enumerate(iris.feature_names):
    print(f"{feature}: Mean={X_test_mean[i]:.3f}, Std={X_test_std[i]:.3f}")
```

```python
# 2. Classify with logistic regression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import StratifiedGroupKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Create pipeline with feature scaling and logistic regression
pipeline = make_pipeline(StandardScaler(), LogisticRegressionCV(cv=5, random_state=42))

# Perform k-fold cross-validation
pipeline.fit(X_train, y_train)

# Check the chosen hyperparameters and model coefficients
print("Chosen hyperparameters: ")
print("Regularization parameter (C):", pipeline.named_steps['logisticregressioncv'].C_)
print("Number of iterations: ", pipeline.named_steps['logisticregressioncv'].n_iter_)

# Check the coefficients of the logistic regression model
print("\nModel coefficients: ")
print(pipeline.named_steps['logisticregressioncv'].coef_)
```

```python
# 3. Evaluate model
from sklearn.metrics import confusion_matrix, classification_report

# Make predictions on test data
y_pred = pipeline.predict(X_test)

# a) Check first 10 values of y_test against prediction
print("Manual comparison of the 10 first values of y_test and predictions")
for true, pred in zip(y_test[:10], y_pred[:10]):
    print(f"True: {true}, Predicted: {pred}")

# b) Plot confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis', cbar=False)
plt.title('Confusion matrix')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()

# c) Print classification report
print("Classification report: ")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
# 4. k-folded cross-validation for evaluation
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import numpy as np

# Define number of folds for cross-validation
n_splits = 5

# Initiate lists to store scores    
precision_scores = []
recall_scores = []
f1_scores = []
accuracy_scores = []

# Initiate KFold cross-validation
kf = KFold(n_splits=n_splits)

# Perform k-fold cross-validation
for train_index, test_index in kf.split(X_train_scaled):
    X_train_cv, X_test_cv = X_train_scaled[train_index], X_train_scaled[test_index]
    y_train_cv, y_test_cv = y_train.iloc[train_index], y_train.iloc[test_index]
    # Fit the model
    pipeline.fit(X_train_cv, y_train_cv)
    # Make predictions
    y_pred_cv = pipeline.predict(X_test_cv)
    # Calculate precision, recall, f1-score and accuracy for each fold
    precision_scores.append(precision_score(y_test_cv, y_pred_cv, average='macro'))
    recall_scores.append(recall_score(y_test_cv, y_pred_cv, average='macro'))
    f1_scores.append(f1_score(y_test_cv, y_pred_cv, average='macro'))
    accuracy_scores.append(accuracy_score(y_test_cv, y_pred_cv))

# Calculate mean scores
mean_precision = np.mean(precision_scores)
mean_recall = np.mean(recall_scores)
mean_f1 = np.mean(f1_scores)
mean_accuracy = np.mean(accuracy_scores)

print("Mean Precision: ", mean_precision)
print("Mean recall: ", mean_recall)
print("Mean f1-score: ", mean_f1)
print("Mean accuracy: ", mean_accuracy)
```
