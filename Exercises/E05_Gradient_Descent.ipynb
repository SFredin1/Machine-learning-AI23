```python
# 0. Simulate dataset
# Set random seed to 42, 
# (1000,2) samples from X~\mu(0,1), i.e 1000 rows and 2 columns
# 1000 samples from \epsilon ~ N(0.1)
# Y = 3x_1 + 5x_2 + 3 + \epsilon, where x_i is the column i of X
import numpy as np
np.random.seed(42)
X = np.random.uniform(0,1,(1000,2))
epsilon = np.random.normal(0,0.1,1000)
Y = 3*X[:,0] + 5*X[:,1] + 3 + epsilon

# Finally add a column of ones for the intercept to X
X = np.c_[np.ones(1000), X]
print(X.shape)
print(Y.shape)
print(X[:5])

```

# $\vec{\theta}$

$\vec{\theta} = (\theta_0, \theta_1, \theta_2)^T$

$\vec{\alpha}$

```python
# 1. Gradient Descent - learning rate
# Use gradient descent to calculate $\vec{\theta} = (\theta_0, \theta_1, \theta_2)^T$
# a) Use n = 0.1 an calculate $\vec{\theta}$ for each fifth epoch from 1 to 500
n = 0.1
epochs = 500
theta = np.random.rand(3)
for epoch in range(epochs):
    gradient = 2*X.T.dot(X.dot(theta) - Y)/1000
    theta = theta - n*gradient
    if epoch % 5 == 0:
        print(f"Epoch {epoch}: {theta}")
    
# plot these $\vec{\theta}$ values against epochs
import matplotlib.pyplot as plt
theta = np.zeros(3)
n = 0.1
epochs = 500
theta_values = []
for epoch in range(epochs):
    gradient = 2*X.T.dot(X.dot(theta) - Y)/1000
    theta = theta - n*gradient
    if epoch % 5 == 0:
        theta_values.append(theta)
theta_values = np.array(theta_values)
plt.plot(range(0,500,5), theta_values[:,0], label='theta0')
plt.plot(range(0,500,5), theta_values[:,1], label='theta1')
plt.plot(range(0,500,5), theta_values[:,2], label='theta2')
plt.legend()
plt.title('Gradient descent, n=0.1')
plt.show()
```

```python
# b) Do the same as for a) but with learning rate, 5000 epochs and for each 20th epoch.
theta = np.zeros(3)
n = 0.1
epochs = 5000
theta_values = []
for epoch in range(epochs):
    gradient = 2*X.T.dot(X.dot(theta) - Y)/1000
    theta = theta - n*gradient
    if epoch % 20 == 0:
        theta_values.append(theta)
theta_values = np.array(theta_values)
plt.plot(range(0,5000,20), theta_values[:,0], label='$theta_0$')
plt.plot(range(0,5000,20), theta_values[:,1], label='$theta_1$')
plt.plot(range(0,5000,20), theta_values[:,2], label='$theta_2$')
plt.title('Gradient descent, n=0.01')
plt.xlabel('Epochs')
plt.ylabel('Theta values')
plt.legend()
plt.show()
```
